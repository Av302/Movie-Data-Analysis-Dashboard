```python
import requests
import pandas as pd
import time

def fetch_movie_data(api_key, year):
    base_url = "http://www.omdbapi.com/"
    movies = []
    page = 1
    
    while True:
        params = {
            'apikey': api_key,
            's': 'movie',  # Search for all movies
            'type': 'movie',
            'y': year,
            'page': page
        }
        response = requests.get(base_url, params=params)
        if response.status_code == 200:
            data = response.json()
            if data.get('Search'):
                movies.extend(data['Search'])
                total_results = int(data['totalResults'])
                print(f"Fetched page {page} for year {year}. Total results: {total_results}")
                if len(movies) >= total_results:
                    break
                page += 1
                time.sleep(1)  # Add a delay to avoid hitting rate limits
            else:
                print(f"No more results for year {year} on page {page}")
                break
        else:
            print(f"Error {response.status_code} fetching data for year {year}, page {page}")
            break
    
    return pd.DataFrame(movies)

# Usage
api_key = ''#your Api key
year = 2021 # Change year 

df = fetch_movie_data(api_key, year)
print(f"Total movies fetched for {year}: {len(df)}")
print(df.head())
```
